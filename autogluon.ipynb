```python
import pandas as pd
```

```python
df = pd.read_csv("bikes_dataset.csv")
df
```

```python
from autogluon.tabular import TabularPredictor

#AutoGluon will recognize this is a multi-class classification task,
#perform automatic feature engineering, train multiple models,
#and then ensemble the models to create the final predictor
predictor = TabularPredictor(label='bikes_hired', eval_metric= 'r2').fit(train_data=df)
```

```python
predictor.leaderboard()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data_raw, test_data_raw = train_test_split(df, test_size=0.2, random_state = 10)
print(f"Train set has: {len(train_data_raw)} rows")
print(f"Test set has: {len(test_data_raw)} rows")
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

#AutoGluon will recognize this is a multi-class classification task,
#perform automatic feature engineering, train multiple models,
#and then ensemble the models to create the final predictor
predictor = TabularPredictor(label='bikes_hired', eval_metric= 'r2').fit(train_data=train_data_raw)
```

```python
#for training data
predictor.leaderboard()
```

```python
#for test data
y_pred = predictor.predict(test_data_raw.drop(columns=['bikes_hired']))
y_pred.head()
```

```python
predictor.evaluate(test_data_raw, silent=True)
```

```python
predictor.leaderboard(test_data_raw)
```
